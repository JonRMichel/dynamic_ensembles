```python
data['YrQ'] = data[['YEAR', 'QUARTER']].apply(lambda s: str(s[0])+str('Q')+str(s[1]) , axis =1)
data['INDUSTRY'] = data['INDUSTRY'].fillna(0)
```

# Dynamic Ensembles

```python
particle_count = 1000
particle_df = pd.DataFrame(columns = range(0,particle_count) )
phi = 20.0

forecast_df = pd.DataFrame( columns = ['YrQ','forecast']).set_index('YrQ')

year = years[0]
forecasters = pd.DataFrame(forecasts_wide.loc[year].dropna()).index.values
particles = random.choices(forecasters, k = particle_count)

for year in years[0:-2]:
    forecast = 0
    forecasters = pd.DataFrame(forecasts_wide.loc[year].dropna()).index.values
    weights = len(particles) * [0]
    
    for particle in range(0,len(particles)):
        if particles[particle] not in forecasters:
            particles[particle] = random.choices(forecasters, k =1)[0]
        
        forecast = forecast + forecasts_wide.loc[year, particles[particle]]
        weights[particle] = np.square(forecasts_wide.loc[year, particles[particle]] - historical.loc[year].values[0])/phi
    
    forecast = forecast/particle_count
    forecast_df.loc[year] = forecast
    
    new_particles = random.choices(particles, weights, k =particle_count)
    particle_df.loc[year] = new_particles

    


#     error_df.loc[year] = forecasts_wide.loc[year]  - historical.loc[year].values[0]
```

```python
fig, ax = plt.subplots(figsize = (16,8))

forecast_df.plot(ax = ax)
historical.plot(ax = ax)
forecasts_wide.mean(axis =1).plot(ax=ax)
```

```python
error_analysis = pd.DataFrame(columns = ['particle_mean','mean'])

for year in years[0:-2]:
    hist = historical.loc[year].values[0]
    error_analysis.loc[year,'particle_mean'] = forecast_df.loc[year,'forecast'] - hist
    error_analysis.loc[year,'mean'] = forecasts_wide.mean(axis = 1).loc[year] - hist
    
```

```python
np.square(error_analysis).diff(axis = 1).plot()
```
